```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
predictions = np.load("predictions.npy")
predictions.shape
```

```python
long_data = predictions.reshape(-1, order='C')
assert np.sum(predictions[0] - long_data[:4096]) == 0

timescale_min=np.linspace(0,60,long_data.shape[0])
```

```python
fig,ax=plt.subplots()
ax.plot(timescale_min, long_data)
ax.grid()
ax.set_xlabel("Time (min)")
ax.set_ylabel("Network prediction")
```
